# **1. GPU 사용하는 방법**

```python
import torch
import torch.nn as nn
```

```python
x = torch.cuda.FloatTensor(2, 2) 
# 에러가 발생할 경우 런타임->런타임 유형변경에서 GPU를 선택
print(x)
# GPU가 여러개 있는 경우 GPU 번호를 확인
# 같은 디바이스에 있는 텐서끼리만 연산이 가능
```

```python
x = torch.FloatTensor(2, 2)
print(x)
```

```python
# GPU로 보냄
x.cuda()
```

```python
# GPU0으로 지정
d = torch.device('cuda:0')
```

```python
# move가 아니라 copy로 됨
x.cuda(device=d)
```

```python
x.device
```

# **2. Convert to CPU tensor from CUDA tensor**

```python
x = torch.cuda.FloatTensor(2, 2)
```

```python
x = x.cpu()
```

```python
d = torch.device('cpu')
x = x.to(d)
```

# **3. Move model from CPU to GPU**

```python
def print_params(model):
  for p in model.parameters():
    print(p)
```

```python
linear = nn.Linear(2, 2)
print_params(linear)
```

```python
linear = linear.cuda()
print_params(linear)
```

```python
linear = linear.cpu()
print_params(linear)
```

```python
d = torch.device('cuda:0')
linear = linear.to(d)
print_params(linear)
```

```python
linear.device # 모델은 device attribute 를 제공하지 않음
```
